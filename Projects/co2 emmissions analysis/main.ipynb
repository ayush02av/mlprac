```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df.columns = ['country', 'year', 'value']
```

```python
plt.figure(figsize = (12, 10))
avg = {}
for country in df.country.unique().tolist():
    df_country = df[df.country == country]
    avg[country] = df_country.value.mean()
    plt.plot(df_country.year, df_country.value, label = country)
plt.show()
```

CO2 emissions on downfall 2005 onwards for almost every country

```python
avg = sorted(avg.items(), key = lambda x: x[1])
```

```python
countries = avg[-5:][::-1]
```

```python
plt.figure(figsize = (12, 10))
for country in countries:
    df_country = df[df.country == country[0]]
    plt.plot(df_country.year, df_country.value, label=country[0])
plt.legend(loc='upper right')
plt.show()
```
