```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
df = df[['Gender', 'Age', 'EstimatedSalary', 'Purchased']]
```

```python
df.isnull().sum()
```

```python
df['Purchased'].value_counts()
```

```python
df.head()
```

```python
x = df.iloc[:, :-1]
```

```python
x.head()
```

```python
x['Gender'].value_counts()
```

```python
x.loc[x['Gender'] == 'Male', 'Gender'] = 1
x.loc[x['Gender'] == 'Female', 'Gender'] = 0
```

```python
x.head()
```

```python
y = df.iloc[:, -1]
```

```python
y.head()
```

```python
model = tf.keras.Sequential()
```

```python
model.add(tf.keras.layers.Dense(1, input_dim = len(x.iloc[0, :]), activation = 'sigmoid'))
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = 'rmsprop', metrics = ['accuracy'])
```

```python
history = model.fit(x.astype(np.int16) , y.astype(np.int16), epochs = 300, verbose = 1)
```

```python
plt.plot(history.history['loss'])
```
